# Задание 1

```python
def square_integrate(function, a: float, b: float, nseg: int = 1000):
    
    dx = (b - a) / nseg
    return dx * (0.5 * (function(a) + function(b)) + 
                 np.sum([function(a + i * dx) for i in range(1, nseg)]))
```

```python
def func1(x):
    return 5 / 6 * x + 0.5 * square_integrate(lambda s: x ** 2 * s, a = 0, b = 1)
```

```python
x, y, dt = handler.make_signal('other', interval = (0, 1), func = lambda x: x)
y_approx = np.array([func1(t) for t in x])
```

```python
display.clear()

display.make_plot(0, x, y, linewidth = 3, legend = 'y_exact')
display.make_plot(0, x, y_approx, color = 'red', linewidth = 5, linestyle = (0, (1, 10)), legend = 'y_approx')

display.plot(title = f'Метод квадратур')
```
